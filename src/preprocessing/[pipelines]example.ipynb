```python
import pandas as pd
from transformers import pipeline
```

```python
df = pd.read_csv('./../../data/original_data/pre_speech2010-2022.csv')
df
```

```python
qa = pipeline('question-answering')
sia = pipeline('sentiment-analysis')
```

```python
df0 = df.copy()
df0['answer'] = ''
df0['score'] = ''
```

```python
for idx in df0.index:
    text = df0.loc[idx, 'speech']
    answer = qa(question='Could you explain the US economy?', context=text)['answer']
    result = sia(answer)[0]
    if result['label'] == 'NEGATIVE':
        score = result['score'] * -1
    else:
        score = result['score']
    print(answer, result)
    df0.loc[idx, 'answer'] = answer
    df0.loc[idx, 'score'] = score
```

```python
df0 = df0[['date', 'year', 'quarter', 'score']]
df0
```

```python
df0.to_csv('./../../data/speech2.csv', index=False)
```
