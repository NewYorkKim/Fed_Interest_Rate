```python
import requests
from bs4 import BeautifulSoup
import os
import time
import pandas as pd
```

```python
beige_book = []
date_list = []

for year in range(1996, 2022+1):
    if year != 2022:
        url = f'https://www.federalreserve.gov/monetarypolicy/beigebook{year}.htm'
    else:
        url = 'https://www.federalreserve.gov/monetarypolicy/beige-book-default.htm'
    res = requests.get(url)
    res.raise_for_status()

    soup = BeautifulSoup(res.text, 'lxml')
    tbody = soup.find('tbody')
    htmls = tbody.find_all('a')
    
    for html in htmls:
        if '.htm' in html['href']:
            url
            try:
                url_a = html['href'] 
                if 'default' not in url_a:
                    res_a = requests.get(url_a)
                    soup_a = BeautifulSoup(res_a.text, 'lxml')
                    div = soup_a.find('div', {'id': 'div_summary'})
                    raw_text = div.find('p').previous_sibling.text
                    ps = div.find_all('p')
                    for p in ps:
                        if p.find('strong'):
                            break
                        elif p.find('b'):
                            break
                        raw_text += p.text 
                    text = raw_text.replace('\n', ' ').replace('\r', '').replace('\t', '').strip()
                    date = soup_a.find('h1', {'class': 'border'}).text.split('- ')[-1]
                else:
                    res_a = requests.get(url_a)
                    soup_a = BeautifulSoup(res_a.text, 'lxml')
                    td = soup_a.find_all('td')[4]
                    ps = td.find_all('p')
                    raw_text = ''
                    for p in ps:
                        if p.find('strong'):
                            break
                        elif p.find('b'):
                            break
                        raw_text += p.text
                    text = raw_text.replace('\n', ' ').replace('\r', '').replace('\t', '').strip()
                    date = soup_a.find('font').text
            except:
                url_a = 'https://www.federalreserve.gov' + html['href']
                res_a = requests.get(url_a)
                soup_a = BeautifulSoup(res_a.text, 'lxml')
                div = soup_a.find('div', {'id': 'article'})
                raw_text = div.find_all('p')[0].text + div.find_all('p')[1].text + div.find_all('p')[2].text
                text = raw_text.replace('\n', ' ').replace('\r', '').replace('\t', '').strip()
                date = soup_a.find('div', {'class': 'page-title'}).text.split(' - ')[-1].strip()
            
            beige_book.append(text)
            date_list.append(date)
    
    print(year)
```

```python
len(beige_book), len(date_list)
```

```python
df = pd.DataFrame([beige_book, date_list]).T
df.columns = ['beige_book', 'date']
df
```

```python
df['date'] = pd.to_datetime(df['date'], infer_datetime_format=True)
df
```

```python
df['year'] = df['date'].dt.year
df
```
