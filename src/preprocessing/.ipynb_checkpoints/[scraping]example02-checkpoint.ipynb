```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
```

```python
url = 'https://www.federalreserve.gov/newsevents/speeches.htm'
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get(url)

element = driver.find_element_by_xpath('/html/body/div[3]/div[2]/div/div[1]/form/div[2]/div/div[1]/span/button')
element.click()
element = driver.find_element_by_xpath('/html/body/div[3]/div[2]/div/div[1]/form/div[2]/div/div[1]/div/ul/li[2]/button[2]')
element.click()
element = driver.find_element_by_xpath('/html/body/div[3]/div[2]/div/div[1]/form/div[2]/div/div[1]/input')
element.click()
element.send_keys('01/01/2010')

time.sleep(1)

element = driver.find_element_by_xpath('/html/body/div[3]/div[2]/div/div[1]/form/div[5]/a')
element.click()

time.sleep(1)
```

```python
#페이지 반복문

speech_list = []
date_list = []

for page in range(33):
    elements = driver.find_elements_by_class_name('ng-binding')
    
    time.sleep(3)
    
    for element in elements:
        link = element.get_attribute('href')
        if link:
            res = requests.get(link)
            soup = BeautifulSoup(res.text, 'lxml')
            block = soup.find('div', {'class': 'col-xs-12 col-sm-8 col-md-8'})
            speech = block.find_all('p')
            text = ''

            if block.find('div', {'class': 'sr-only'}) != None:
                for p in speech:
                    if p in block.find('div', {'class': 'sr-only'}):
                        continue
                    else:
                        if p.text == 'Note' or p.text == 'References':
                            break
                        elif 'Return to text' in p.text:
                            break
                        text += ' ' + p.text.replace('\n', ' ')
                date = soup.find('p', {'class': 'article__time'}).text
        
                speech_list.append(text)
                date_list.append(date)
            
            elif block.find('div', {'class': 'panel-body'}) != None:
                continue
                
            else:
                for p in speech:
                    if p.text == 'Note' or p.text == 'References':
                        break
                    elif 'Return to text' in p.text:
                        break
                    text += ' ' + p.text.replace('\n', ' ')
            
                date = soup.find('p', {'class': 'article__time'}).text
            
                speech_list.append(text)
                date_list.append(date)
    print(page + 1)

    next = driver.find_element_by_css_selector('#article > ul.hidden-xs.ng-untouched.ng-valid.ng-isolate-scope.pagination.ng-not-empty.ng-dirty > li.pagination-next.ng-scope > a')
    next.click()
    
print('End')
```

```python
s = pd.DataFrame(speech_list)
s
```

```python
df = pd.DataFrame([speech_list, date_list]).T
df.columns = ['speech', 'date']
df
```

```python
df['date'] = pd.to_datetime(df['date'], infer_datetime_format=True)
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['quarter'] = df['date'].dt.quarter
df
```

```python
df.to_csv('./../../data/original_data/speech2010-2022.csv', index=False)
```
